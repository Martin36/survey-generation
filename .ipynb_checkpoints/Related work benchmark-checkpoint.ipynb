```python
import seaborn as sns
import pandas as pd

from utils_package.util_funcs import load_json

STATS_PATH = "stats/rel_work_benchmark_stats.json"

stats = load_json(STATS_PATH)

penguins = sns.load_dataset("penguins")
penguins
```

```python
stats
```

```python
def convert_to_df(stats: dict):
    data = {"dataset": [], "length": []}
    for dataset in stats:
        if "total" in dataset:
            continue
        dataset_name = dataset.split()[0]
        dist = stats[dataset]
        for key in dist:
            val = dist[key]
            for i in range(val):
                data["dataset"].append(dataset_name)
                data["length"].append(int(key))
    df = pd.DataFrame.from_dict(data)
    return df 
```

```python
df = convert_to_df(stats)
df
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.histplot(data=df, x="length", hue="dataset", binwidth=200, binrange=(0,20000))
```

```python
longer_than_4096 = df.loc[df["length"] >= 4096]
percentage = 100*(len(longer_than_4096)/len(df))
print(f"{len(longer_than_4096)} out of {len(df)} are longer than 4096 e.g. {percentage}%")
```
